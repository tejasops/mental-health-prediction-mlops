# 02 - Data Preprocessing
Clean and preprocess the survey data, then save to data/processed/.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from src.data_loader import load_raw_data
from src.preprocessing import (
    drop_unnecessary_columns, fill_missing_values,
    clean_gender, clean_age, clean_self_employed,
    clean_work_interfere, encode_features, run_preprocessing
)
```

```python
df = load_raw_data()
print(f'Raw shape: {df.shape}')
df.head()
```

## Step-by-step preprocessing

```python
df = drop_unnecessary_columns(df)
print(f'After dropping columns: {df.shape}')
df.head()
```

```python
df = fill_missing_values(df)
print(f'Missing values remaining: {df.isnull().sum().sum()}')
```

```python
df = clean_gender(df)
print(f'Unique genders: {df["Gender"].unique()}')
print(f'Shape after gender cleaning: {df.shape}')
```

```python
df = clean_age(df)
print(f'Age range: {df["Age"].min()} - {df["Age"].max()}')
print(f'Age range bins: {df["age_range"].unique()}')
```

```python
df = clean_self_employed(df)
df = clean_work_interfere(df)
df = df.drop(columns=['Country'])
print(f'Final columns: {list(df.columns)}')
```

```python
df, label_dict, encoders = encode_features(df)
print('Label mappings:')
for k, v in label_dict.items():
    print(f'  {k}: {v}')
```

```python
df.to_csv('../data/processed/survey_cleaned.csv', index=False)
print(f'Saved processed data: {df.shape}')
df.head()
```
