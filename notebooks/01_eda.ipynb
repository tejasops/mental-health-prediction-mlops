# 01 - Exploratory Data Analysis
Explore the Mental Health in Tech survey dataset.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_loader import load_raw_data

%matplotlib inline
sns.set(style='whitegrid')
```

```python
df = load_raw_data()
print(f'Shape: {df.shape}')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Missing Data

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

## Gender Distribution

```python
print(f'Unique genders: {df["Gender"].nunique()}')
df['Gender'].value_counts().head(20)
```

## Age Distribution

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], bins=50, kde=True)
plt.title('Age Distribution (Raw)')
plt.xlabel('Age')
plt.show()
```

## Treatment Distribution

```python
plt.figure(figsize=(6, 4))
df['treatment'].value_counts().plot(kind='bar')
plt.title('Treatment Distribution')
plt.ylabel('Count')
plt.show()
```

## Country Distribution

```python
plt.figure(figsize=(14, 6))
df['Country'].value_counts().head(15).plot(kind='bar')
plt.title('Top 15 Countries')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Treatment by Family History

```python
pd.crosstab(df['family_history'], df['treatment']).plot(kind='bar', figsize=(8, 5))
plt.title('Treatment by Family History')
plt.ylabel('Count')
plt.show()
```
